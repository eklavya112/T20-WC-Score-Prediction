# Data Extraction

```python
#importing the libraries
import os
import numpy as np
import pandas as pd
from yaml import safe_load              #for yaml to pandas
from tqdm import tqdm
```

```python
# with the help of os module let's extract all filenames and store them in a list
filenames=[]
for file in os.listdir("t20s"):
    filenames.append(os.path.join("t20s",file))
```

```python
#let's check whether filenames contains all names of files with their directory
filenames[:10]
```

Now we will take each file individually and then after applying json normalizer to it then we will convert it to pandas dataframe with safeload. And we are adding one more column with match id like match no. 1,2 etc.

```python
dataframe = pd.DataFrame()
count=1
for file in tqdm(filenames):
    with open(file,'r') as data:
        df = pd.json_normalize(safe_load(data))
        df['match_id']=count
        dataframe=dataframe.append(df)
        count+=1

return dataframe
```

```python
matches.isnull().sum()
```

```python
# Code to extact ball by ball data from innings column and add it to dataframe 
count=0
df_delivery=pd.DataFrame()
for index,row in matches.iterrows():
    #if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
       # count+=1
       # continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_dismissed.append('0')
                
    df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_dismissed,
            'city':city,
            'venue':venue
        })
    df_delivery = df_delivery.append(df)
```

So now we have data for ball by by for the first innings and we can see that we have batting team in it, so we can make another column with bowling team and for that we have to extract the bowling team from the dataset

```python
# function to get bowling team column
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
df_delivery['bowling_team'] = df_delivery.apply(bowl,axis=1)
```

City is a categorical data and there are 87 cities listed in data so we won't consider all. We can start considering those cities where almost 5 matches ahve been played. 5 matches first innings means atleast 600 balls.

```python
eligible_cities = output['city'].value_counts()[output['city'].value_counts() > 600].index.tolist()
```

```python
output = output[output['city'].isin(eligible_cities)]
```

```python
output
```

No we have to calculate current score means what was the score after a certain ball

```python
output['current_score'] = output.groupby('match_id').cumsum()['runs']
```

```python
output
```

```python
output['over'] = output['ball'].apply(lambda x:str(x).split(".")[0])
output['ball_no'] = output['ball'].apply(lambda x:str(x).split(".")[1])
output
```

```python
output['balls_bowled'] = (output['over'].astype('int')*6) + output['ball_no'].astype('int')
output
```

```python
output['balls_left'] = 120 - output['balls_bowled']
output['balls_left'] = output['balls_left'].apply(lambda x:0 if x<0 else x)
output
```

As we want to know how many wickets are left after a certain period or overs. For that we need to extract this data from above dataset as in the player_dismiised column it is given as we can see that if a wicket down than that **players name is given or else the value is 0**. So with it we will extract wickets left data with the following code

```python
output['player_dismissed'] = output['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
output['player_dismissed'] = output['player_dismissed'].astype('int')
output['player_dismissed'] = output.groupby('match_id').cumsum()['player_dismissed']
output['wickets_left'] = 10 - output['player_dismissed']
```

```python
output
```

We want Current run rate also as one of our feauture so we will extract that also and add that current run rate **(crr)** column into our dataset

```python
output['crr'] = (output['current_score']*6)/output['balls_bowled']
```

```python
output
```

**Cricket is a game of momentum** and to predict something in cricket specially score, it is very important to know where the momentum is going in terms of runs, wickets or even in fielding. Here we are using last three overs runs to know the momentum. We can also use 2 overs or even 5 overs. We will do that with the help of following code.

```python
groups = output.groupby('match_id')

match_ids = output['match_id'].unique()
last_three = []
for id in match_ids:
    last_three.extend(groups.get_group(id).rolling(window=18).sum()['runs'].values.tolist())  
    # 18 because 3 overs means 18 balls 
```

```python
output['last_three'] = last_three
output
```

Now what we will see is that first 3 overs or 18 balls of all mathces will have NaN value because we are adding last three overs data so that will be available after 3 overs only. **Now again from the above data we will extract all the relevant columns we want in our model**.Our target is to get total runs at the end of the innings so that should also be there in our dataset.

```python
#code for total runs
final_df = output.groupby('match_id').sum()['runs'].reset_index().merge(output,on='match_id')
```

```python
final_df
```

runs_x in the above dataset is showing total runs at the end of innings and runs_y is showing run in that particular bowl

```python
#All the columns or data that i am using for my model
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_three','runs_x']]
```

```python
final_df.isnull().sum()
```

```python
#remove rows with missing data
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

We should try from our end that there is no bias developed due to our data extraction and analysis work. So for that we will **Suffle the data**.

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

Finally we are at the door of model building as we have done data extraction and feauture extraction as well. Now further tasks are to split data into training and testing dataset and build model and do the prediction.

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
# Because we have three columns in our dataset are string so we need column transformer
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
# Now we are creating a sklearn pipeline where first data will transform as per trf then it will we scaled and finally xgboost
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
#training the model and seeing mae and r^2 score
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

So it is giving good r^2 value and we can save it so will save the model so that we can deploy it.
